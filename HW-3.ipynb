##### a) What is the pdf of the sum of two identical exponential distributions?

$$f(x) = \lambda e^{-\lambda x}$$
$$f(x)*f(x) = \int_{0}^{x}f(x-z)f(z)dz$$
$$\int_{0}^{x} e^{-\lambda(x-z)}e^{-\lambda z}dz = \int_{0}^{x}e^{-\lambda x}*e^{\lambda z}*e^{-\lambda z}dz = \int_{0}^{x}e^{-\lambda x}dz$$


$$f(x)*f(x) = ze^{-\lambda x}\bigg|_{z = 0}^{z = x}$$

$$f(x)*f(x) = xe^{-\lambda x}$$

This results in a non-normalized gamma distribution with parameter $\alpha$ = 2

##### b) What is the pdf of the ratio of two zero-mean unity variance normal distributions  $X_1/X_2$ ?

Let $P_{X_1}(x)$ and $P_{X_2}(y)$ be two zero mean normal distributions with variance = 1, then:

$$\int_{0}^{\infty}ye^{-ay^2}dy = \frac{1}{2a}$$

$|y|e^{\frac{-y^2\left(z^2 + 1\right)}{2}}$ is symmetric about the origin. Therefore, over the whole interval the integral equals $\frac{1}{a}$ where $a = \frac{z^2+1}{2}$

$$P_z(z)=\frac{1}{2\pi}\frac{1}{\frac{z^2+1}{2}} = \frac{1}{\pi\left(z^2 + 1\right)}$$

This is a Cauchy distribution with parameters: $\gamma = 1$ and $x-x_0=z$

##### c) The 2D Gaussian dsitribution looks like a smeared spot. Mathematically, this is given by $X \hat{i} + Y \hat{j}$ where both X and Y are from 1D Gaussian distributions. If we measure the amplitude of this vector, what is its pdf?

Let $Z = \sqrt{X^2 + Y^2}$ where $P_X = \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$ and $P_Y = \frac{1}{\sqrt{2\pi}}e^{-\frac{y^2}{2}}$

$z = \sqrt{x^2 + y^2}$

Then the cdf of two independent orthonormal normally distributed random variables is:

$$C_Z=\int\int_{D_z}P_X(x)P_Y(y)dA$$

$$\frac{1}{2\pi}\int\int_{D_z}e^{\frac{-1}{2}(x^2+y^2)}dA = \frac{1}{2\pi}\int\int_{D_z}e^{\frac{-1}{2}z^2}dA$$

$$C_z=\frac{1}{2\pi}\int_0^{2\pi}\int_0^ze^{\frac{-1}{2}r^2}rdrd\theta = \int_0^zre^{\frac{-1}{2}r^2}dr$$

The pdf is the derivative of the cdf, so $P_z$ is:

$$P_z = ze^{-\frac{z^2}{2}}$$

Where z is a real number between 0 and infinity. This is equivalent to a Rayleigh distribution with parameter $\sigma = 1$
